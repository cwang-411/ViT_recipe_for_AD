```python
def check_magnetic_field_strength(nii_file):
    img = nib.load(nii_file)
    header = img.header
    if 'pixdim' in header:
        pixdim = header['pixdim']
        if len(pixdim) > 4:
            magnetic_field_strength = pixdim[4]
            if magnetic_field_strength == 1.5:
                return "MRI volume acquired at 1.5T"
            elif magnetic_field_strength == 3.0:
                return "MRI volume acquired at 3T"
            else:
                return "Unknown magnetic field strength"
        else:
            return "Magnetic field strength information not found in header"
    else:
        return "Magnetic field strength information not found in header"

# Example usage:
for sample_path in list_of_t1s:
    result = check_magnetic_field_strength(sample_path)
    print(result)
```
